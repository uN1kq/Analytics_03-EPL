```python
# Importing the library
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing dataset
data = pd.read_csv('EPL_20_21.csv')
data.head()
```

```python
#Data about the data
data.info()
data.describe()
```

```python
#Checking for the null/empty data
data.isnull().sum()
```

```python
#Creating two new columns
data['MinsPerMatch'] = (data['Mins']/data['Matches']).astype(int)
data['GoalsPerMatch'] = (data['Goals']/data['Matches']).astype(float)
data.head()
```

```python
#Total Number of Goals in the entire season
total_goals = data['Goals'].sum()
print("Total no.of Goals in the season 2020-2021:- ",total_goals)
```

```python
#Total Number of Penalty Goals in the entire season
total_penalties = data['Penalty_Goals'].sum()
print("Total no.of Penalties in the season 2020-2021:- ",total_penalties)
```

```python
#Total Number of Penalty Attempted in the entire season
penalties_attempt = data['Penalty_Attempted'].sum()
penalties_attempt
```

```python
#Pie-Chart of Penalties Scored vs Not Scored
plt.figure(figsize=(14,8))
not_scored = penalties_attempt - total_penalties
dat = [not_scored, total_penalties]
labels = ['Not scored','Scored']
color = sns.color_palette("Set2")
myexplode = [0.2, 0]
plt.pie(dat, labels= labels, colors=color,autopct = "%.02f", explode = myexplode)
plt.show()
```

```python
#Most players from a country 
nations = data.groupby("Nationality").size().sort_values(ascending=False)
nations.head(10).plot(kind = "bar",figsize =(10,8),xlabel = "Country",ylabel = "No.of Players",)



### 
# data["Nationality"].value_counts().nlargest(10)
```

```python
#Clubs with max player sin their squad
data['Club'].value_counts().nlargest(10).plot(kind='bar',figsize =(12,8),color = sns.color_palette("rocket"))
```

```python
#Clubs with min player sin their squad
data['Club'].value_counts().nsmallest(10).plot(kind='bar',figsize =(12,8),color = sns.color_palette("rocket"))
```

```python
#Players based on age group
under_19 = data[data['Age'] < 20]
age_20_28 = data[(data['Age'] > 20) & (data['Age'] < 28)]
above_29 = data[data['Age'] >= 29]
age_data = [under_19['Name'].count(), age_20_28['Name'].count(), above_29['Name'].count()]
lab = ['<19','20 & 28','>29']
plt.pie(age_data, labels=lab, autopct = '%.1f%%')
plt.show()
```

```python
#Under 19 player in each club
under_19['Club'].value_counts().plot(kind='bar',figsize=(12,6),color = sns.color_palette('Set2'))
```

```python
#Under 19 players in ManUnited
under_19[under_19['Club'] == 'Manchester United']
```

```python
#Total assists from each club
data.groupby('Club')['Assists'].sum().sort_values(ascending = True).plot(kind='bar',figsize=(20,6),color=sns.color_palette('Set1'))
```

```python
#Top 10 players with max assists
data[['Name','Club','Age','Assists','Matches']].nlargest(n=10, columns='Assists')
```

```python
#Top 10 players with max goals
top_10_GOALS = data[['Name','Club','Age','Goals','Matches']].nlargest(n=10, columns='Goals')
top_10_GOALS
```

```python
#Goals with and without assists
assista = data['Assists'].sum()
da = [total_goals - assista,assista]
plt.pie(da, colors=sns.color_palette('Set2'),autopct='%.2f%%')
plt.show()
```
